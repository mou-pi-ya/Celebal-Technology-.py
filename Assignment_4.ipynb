```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic = sns.load_dataset('titanic')

# Display basic info
print(titanic.shape)
titanic.head()
```

```python
titanic.info()

titanic.describe(include='all')
```

```python
missing = titanic.isnull().sum()
missing = missing[missing > 0]
missing_percent = (missing / len(titanic)) * 100

missing_df = pd.DataFrame({'Missing Values': missing, 'Percentage': missing_percent})
print(missing_df)

plt.figure(figsize=(10, 6))
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
categorical_cols = titanic.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=titanic, palette='pastel')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
numerical_cols = titanic.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(titanic[col].dropna(), kde=True, bins=30, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
for col in ['age', 'fare']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=titanic[col], color='lightcoral')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset
titanic_corr = sns.load_dataset('titanic')

# Keep only numeric columns, and handle NaNs
titanic_corr_clean = titanic_corr.select_dtypes(include=[np.number]).dropna()

# Calculate correlation matrix
corr_matrix = titanic_corr_clean.corr(method='spearman')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Columns Only)", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (sample)
titanic_corr = sns.load_dataset('titanic')

# Encode categorical columns
for col in titanic_corr.columns:
    if titanic_corr[col].dtype == 'object' or str(titanic_corr[col].dtype) == 'category':
        titanic_corr[col] = pd.factorize(titanic_corr[col])[0].astype(int)

# Drop rows with missing values
titanic_corr = titanic_corr.dropna()

# Calculate Spearman correlation
corr_matrix = titanic_corr.corr(method='spearman')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (All Encoded Columns)", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set2')
plt.title('Survival Count by Sex')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=titanic, palette='Set3')
plt.title('Survival Count by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=titanic, x='age', hue='survived', kde=True, multiple='stack')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='pclass', y='age', hue='survived', data=titanic, split=True, palette='muted')
plt.title('Survival by Age and Class')
plt.show()
```
